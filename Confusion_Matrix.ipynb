```python
import pandas as pd
import os

def get_csv_files(directory):
    files = os.listdir(directory)
    return [file for file in files if file.endswith('.csv')]

# Get the list of CSV files
file_names = get_csv_files('inference_output/')
print(file_names)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

def plot_save_confusion_matrix(file_name):
  file_path = 'inference_output/' + file_name
  # Read the CSV file into a DataFrame
  df = pd.read_csv(file_path)

  # Define a function to map the values
  def map_values(x):
      if x in [0, 1]:
          return 0
      else:
          return 1

  # Apply the function to column 'A' to create column 'B'
  df['mapped_gt'] = df['gt'].apply(map_values)
  df['mapped_pred_label'] = df['pred_label'].apply(map_values)

  # Display the DataFrame
  print(df)

  # Compute the confusion matrix
  cm = confusion_matrix(df['mapped_gt'], df['mapped_pred_label'])

  # Display the confusion matrix
  print(cm)

  # Plot the confusion matrix
  plt.figure(figsize=(8, 6))
  # Custom labels
  custom_labels = ['closed-ended', 'open-ended']
  sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=custom_labels, yticklabels=custom_labels)
  plt.xlabel('Predicted Answer')
  plt.ylabel('Question Type')
  plt.title('Confusion Matrix')

  plt.savefig('inference_output/confusion_matrix/' + file_name.replace('.csv', '.png'))
  plt.close()
```

```python
for file in file_names:
  plot_save_confusion_matrix(file)
```
